```python
assets = [json.loads(data['assets'][i]) for i in range(len(data['assets']))]
eq_and_liab = [json.loads(data['eq_and_liab'][i]) for i in range(len(data['eq_and_liab']))]
prices_change = list(data['year_value_change'])
```

```python
# define and fit the final model
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam')

#only assets used
model.fit(assets, prices_change, epochs=1000, verbose=0)
```

```python
# make a prediction
ynew = model.predict(Xnew)
```
